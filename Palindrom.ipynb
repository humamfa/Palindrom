```python
txt = pd.read_csv('numbers.txt', sep=",", names=['Binary', 'Palindrom?'])
df = txt
df['Decimal'] = ""
print(df)
```

```python
pal = []
dec = []

def Palindrom(x):
  rev = ""
  for i in x:
      rev = i + rev
  if x == rev:
    return "Ja"
  else:
    return "Nein"

for i in df['Binary']:
  temp = 0
  num = str(i)
  check = Palindrom(num)
  pal.append(check)
  if check == "Ja":
    multi = len(num)
    for j in num:
      multi -= 1
      if j == "1":
        temp += 2**multi
  dec.append(temp)

df['Palindrom?'] = pal
df['Decimal'] = dec

pd.set_option('display.max_rows', None)
print(df)
```

```python
total10 = 0

for i in range(0, 9):
  total10 += df.loc[i, 'Decimal']

total = 0

for i in df['Decimal']:
  total += i

print('Dezimaler Gesamtwert der ersten 10 Einträge: ' + str(total10))
print('Dezimaler Gesamtwert aller Einträge: ' + str(total))
```
